```python
import pandas as pd
import numpy as np
```

```python
name_list = ["abhi", "surya", "ravi", "himanshu", "gopal", "deepu", "sunil", "mohan", "abhishek", "atul", "lokendra", "yogesh", "nikhil","chetan","sunny"]
```

```python
city_list = ["jaipur","delhi","goa","mumbai"]
```

```python
data = {
    'Customer_ID' : range(1,1001),
    'Name' : [np.random.choice(name_list)
                       for _ in range(1000)],
    'Age' : [np.random.randint(18,99)
                     for _ in range(1000)],
    'City' : [np.random.choice(city_list) 
               for _ in range(1000)],
    "Amount": [np.random.randint(100,1000)
               for _ in range(1000)],
    "Rating": [np.random.uniform(1,10)
               for _ in range(1000)]
}

df = pd.DataFrame(data)

df.to_csv("Syntheic_Customer_Data_Generator.csv", index=False)
print("file 'Syntheic_Customer_Data_Generator.csv' ban gayi hai")
```
